```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dxdt(t, x):
    return 10*x+11*t-5*pow(t, 2)-1


def x(t):
    return pow(t, 2)/2-t


def RK4(x0, t0, tmax, h):
    ts = []
    xs = []
    ts.append(t0)
    xs.append(x0)
    i = 0
    while ts[i]+h <= tmax:
        "Apply Runge Kutta Formulas to find next value of y"
        k1 = dxdt(ts[i], xs[i])
        k2 = dxdt(ts[i] + 0.5 * h, xs[i] + 0.5 * h * k1)
        k3 = dxdt(ts[i] + 0.5 * h, xs[i] + 0.5 * h * k2)
        k4 = dxdt(ts[i] + h, xs[i] + h * k3)

        xs.append(xs[i] + (h / 6)*(k1 + 2 * k2 + 2 * k3 + k4))
        ts.append(ts[i] + h)
        i += 1
    return ts, xs
```

```python
h = pow(2, -8)
t_numerical, x_numerical = RK4(0, 0, 3, h)
t_exact = np.linspace(0, 3, len(x_numerical))
x_exact = x(t_exact)
```

```python
plt.figure(figsize=[13, 8])
plt.plot(t_numerical, x_numerical, label='numerical sol')
plt.plot(t_exact, x_exact, label='exact sol')
plt.title('x(0) = 0')
plt.legend(loc='best')
plt.show()
```

```python
err = np.log(abs(x_numerical-x_exact))
xslength = len(err)
plt.figure(figsize=[13, 8])
plt.scatter(np.arange(xslength), err, label='numerical sol')
# plt.plot(ts_exact, xs_exact, label='exact sol')
plt.legend(loc='best')
plt.show()
```
